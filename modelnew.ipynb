```python
#optional
import cv2
import torch
import torch.nn.functional as F
from PIL import Image
from transformers import AutoFeatureExtractor, AutoModelForImageClassification
from datetime import datetime
import pygame
import csv
import os

# 💽 Init pygame mixer for alert sound
pygame.mixer.init()
pygame.mixer.music.load("siren-alert-96052.mp3")

# 🔍 Load model and extractor
model_name = "google/vit-base-patch16-224"
feature_extractor = AutoFeatureExtractor.from_pretrained(model_name)
model = AutoModelForImageClassification.from_pretrained(model_name)
model.eval()

# 🎯 Custom label mapping and filtering
custom_labels = {
    "comic book": "Anime Character",
    "soccer ball": "Football",
    "volleyball": "Sports Ball"
}
TARGET_LABELS = ["person", "cat", "dog", "knife", "fire", "gun", "animal", "Anime Character"]

alert_classes = ["weapon", "gun", "rifle", "knife", "fire", "explosion", "Anime Character"]

# 📂 Setup log file
log_filename = "detection_log.csv"
if not os.path.exists(log_filename):
    with open(log_filename, mode="w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Timestamp", "Detected Labels"])

# 🎥 Start camera
cap = cv2.VideoCapture(0)
print("📡 Surveillance system started... Press 'q' to quit.\n")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    img_pil = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)).convert("RGB")
    inputs = feature_extractor(images=img_pil, return_tensors="pt")

    with torch.no_grad():
        outputs = model(**inputs)
        probs = F.softmax(outputs.logits, dim=-1)
        topk = torch.topk(probs, k=5)

    results = []
    for i in range(5):
        idx = topk.indices[0][i].item()
        conf = topk.values[0][i].item()
        label = model.config.id2label[idx]
        mapped_label = custom_labels.get(label, label)
        results.append((mapped_label, conf))

    # 🧠 Filter only target labels
    filtered_results = [
        (label, conf) for label, conf in results if label.lower() in [t.lower() for t in TARGET_LABELS]
    ]

    # ⏰ Timestamp
    timestamp = datetime.now().strftime('%H:%M:%S')

    alert = False
    labels_text = []

    if filtered_results:
        for label, conf in filtered_results:
            labels_text.append(f"{label} ({conf * 100:.1f}%)")
            if label.lower() in alert_classes:
                alert = True
    else:
        labels_text.append("No relevant objects detected")

    # 🖼 Draw on frame
    status = "🚨 ALERT" if alert else "✅ SAFE"
    cv2.putText(frame, f"{timestamp} | {status}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7,
                (0, 0, 255) if alert else (0, 255, 0), 2)

    for i, txt in enumerate(labels_text):
        cv2.putText(frame, txt, (10, 60 + i * 25), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)

    # 📸 Save alert snapshot
    if alert:
        filename = f"alert_{datetime.now().strftime('%Y%m%d_%H%M%S')}.jpg"
        cv2.imwrite(filename, frame)
        print(f"📸 Snapshot saved: {filename}")
        if not pygame.mixer.music.get_busy():
            pygame.mixer.music.play()

    # 📝 Log detection
    with open(log_filename, mode="a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([timestamp, "; ".join([f"{label} ({conf:.2%})" for label, conf in filtered_results])])
        print("📝 Prediction logged to detection_log.csv")

    # 🎥 Show live feed
    cv2.imshow("🔍 Smart Surveillance", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import torch
import torch.nn.functional as F
from PIL import Image
from transformers import AutoFeatureExtractor, AutoModelForImageClassification
from datetime import datetime
import pygame
import pyttsx3
from collections import Counter
import csv
import os
from twilio.rest import Client
import gradio as gr

# Twilio setup
account_sid = ''
auth_token = ''
twilio_phone_number = ''
receiver_phone_number = ''
client = Client(account_sid, auth_token)

def send_sms(message):
    message = client.messages.create(
        body=message,
        from_=twilio_phone_number,
        to=receiver_phone_number
    )
    print(f"SMS sent: {message.sid}")

# Audio setup
pygame.mixer.init()
pygame.mixer.music.load("siren-alert-96052.mp3")
engine = pyttsx3.init()

# Load model
model_name = "google/vit-base-patch16-224"
feature_extractor = AutoFeatureExtractor.from_pretrained(model_name)
model = AutoModelForImageClassification.from_pretrained(model_name)
model.eval()

# Custom labels and alert classes
custom_labels = {
    "comic book": "Anime Character", "soccer ball": "Football",
    "volleyball": "Sports Ball", "meat cleaver": "Knife", "cleaver": "Knife",
    "chopper": "Knife", "letter opener": "Knife", "paper knife": "Knife",
    "paperknife": "Knife", "Band Aid": "First Aid", "syringe": "Injection"
}
alert_classes = ["weapon", "gun", "rifle", "knife", "fire", "explosion", "anime character"]

# Face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# CSV setup
csv_file = "surveillance_log.csv"
if not os.path.exists(csv_file):
    with open(csv_file, mode='w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(["Timestamp", "Labels", "Alert", "Faces"])

alert_counter = 0
snap_counter = 0
all_detected_labels = []

def process_frame():
    global alert_counter, snap_counter, all_detected_labels

    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    cap.release()

    if not ret:
        return None, "Camera Error", "❌"

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 5)
    face_count = len(faces)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 200, 100), 2)

    img_pil = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)).convert("RGB")
    inputs = feature_extractor(images=img_pil, return_tensors="pt")

    with torch.no_grad():
        outputs = model(**inputs)
        probs = F.softmax(outputs.logits, dim=-1)
        topk = torch.topk(probs, k=3)

    alert = False
    labels_text = []
    detected_now = []

    for i in range(3):
        idx = topk.indices[0][i].item()
        conf = topk.values[0][i].item() * 100
        label = model.config.id2label[idx].lower()
        mapped_label = next((v for k, v in custom_labels.items() if k in label), label)

        labels_text.append(f"{mapped_label} ({conf:.1f}%)")
        detected_now.append(mapped_label)
        all_detected_labels.append(mapped_label)

        if mapped_label.lower() in alert_classes:
            alert = True

    timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    status = "🚨 ALERT" if alert else "✅ SAFE"
    color = (0, 0, 255) if alert else (0, 255, 0)

    cv2.rectangle(frame, (5, 5), (635, 475), color, 4)
    cv2.putText(frame, f"{timestamp} | {status}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

    for i, txt in enumerate(labels_text):
        cv2.putText(frame, txt, (10, 60 + i * 25), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)

    # Log to CSV
    with open(csv_file, mode='a', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow([timestamp, ", ".join(detected_now), status, face_count])

    if alert:
        alert_counter += 1
        snap_counter += 1
        snap_filename = f"alert_{datetime.now().strftime('%Y%m%d_%H%M%S')}_{snap_counter}.jpg"
        cv2.imwrite(snap_filename, frame)

        send_sms(f"🚨 Alert! Suspicious item detected at {timestamp}. Check the snapshot.")
        if not pygame.mixer.music.get_busy():
            pygame.mixer.music.play()
            engine.say("🚨 Alert! Suspicious item detected!")
            engine.runAndWait()

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    return frame_rgb, ", ".join(labels_text), status

def summary():
    summary_text = f"Total alerts triggered: {alert_counter}\n"
    if all_detected_labels:
        common = Counter(all_detected_labels).most_common(5)
        summary_text += "Top detected labels:\n"
        for label, count in common:
            summary_text += f"• {label}: {count} times\n"
    return summary_text

iface = gr.Interface(
    fn=process_frame,
    inputs=None,
    outputs=[
        gr.Image(label="Live Feed"),
        gr.Textbox(label="Detected Labels"),
        gr.Textbox(label="Alert Status")
    ],
    live=True,
    title="🔍 Smart AI Surveillance",
)

iface.launch()
```
